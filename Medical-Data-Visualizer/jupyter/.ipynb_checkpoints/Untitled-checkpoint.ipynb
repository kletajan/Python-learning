```python
import pandas as pd
# import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import data
df = pd.read_csv('medical_examination.csv')
df.head()
```

```python
# Add 'overweight' column
df['overweight'] = np.where(df['weight'] / (df['height'] * df['height'] / 10000) > 25, 1, 0)
df.head()
```

```python
# Normalize data by making 0 always good and 1 always bad. If the value of 'cholesterol' or 'gluc' is 1, make the value 0. 
#If the value is more than 1, make the value 1.
df['cholesterol'] = np.where(df['cholesterol'] > 0, 1, 0)
df['gluc'] = np.where(df['gluc'] > 0, 1, 0)
df
```

```python
# Create DataFrame for cat plot using `pd.melt` using just the values from 'cholesterol', 'gluc', 'smoke', 'alco', 'active', and 'overweight'.
df.melt(id_vars=['cholesterol','gluc', 'smoke', 'alco', 'active', 'overweight'])
```
