```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
```

```python
# Import data (Make sure to parse dates. Consider setting index column to 'date'.)
df = pd.read_csv('fcc-forum-pageviews.csv', parse_dates=['date'])
```

```python
len(df)
```

```python
# Clean data
minusdata = np.floor(len(df) * 0.025).astype(int)
df = df.drop(range(0, minusdata + 1))
df = df.drop(range(len(df) - minusdata, len(df)))
len(df)
```

```python
df.rename(columns={'date': 'Date', 'value': 'Page Views'}, inplace=True)
df
```

```python
df.plot.line(x="Date", y='Page Views', title="Daily freeCodeCamp Forum Page Views 5/2016-12/2019", figsize=(15,8), ylim=(20000, 200000))
```
